```python
from datetime import timedelta

def ans(duration_ms):
    t = timedelta(milliseconds=duration_ms)
    minutes = t.seconds // 60
    seconds = t.seconds % 60
    return f"{minutes} : {seconds}"
```

```python
spotify['duration_min'] = spotify['duration_ms'].apply(ans)
spotify.head(3)
```

Change in rankings of songs over time to identify trends and patterns.

```python
musical_features = ["danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness",
                    "instrumentalness", "liveness", "valence", "tempo", "time_signature"]

# Compute the correlation matrix of these columns
corr = spotify[musical_features].corr()

# Plot the heatmap of the correlation matrix
sns.heatmap(corr, annot=True)
```
